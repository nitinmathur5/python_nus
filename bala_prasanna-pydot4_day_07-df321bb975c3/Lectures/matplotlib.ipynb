# 1. Simple line plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure() # fig is a container that contains all objects like axes, graphics, labels 
ax = plt.axes()    # axes is the axes in the figure
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));         # this figure is created as background like axes
```

```python
plt.plot(x, np.sin(x)); # call plot function to plot the sine curve
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x)); # create multiple curves
```

adjust color of lines

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

adjust the style of lines

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

combine styles and colors 

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

adjust the axes limmits

```python
plt.plot(x, np.sin(x)) # plot the sine curve

plt.xlim(-1, 11)       # set the domain (-1, 11)
plt.ylim(-1.5, 1.5);   # set the range (-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]); # using the single function to set the axes limits [xmin, xmax, ymin,ymax ]
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');     # using axis function to tighten the plot
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');      # make the length of the x-axis and y-axis equal
```

# 3. Multiples sub plots

```python
ax1 = plt.axes()  # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2]) # these number represents the 
                                       #[left, bottom, width, height] in the figure coordinate system
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-1.2, 1.2)) # set the xticktables none
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))                 # set the range to (-1.2, 1.2)

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)                # 2 rows, 3 colums, ith figure
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)  # plt.subplots_adjust can be used to adjust the spacing between these plots
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3) # set the grid 

plt.subplot(grid[0, 0])  # set four different subplots
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
plt.style.use('seaborn-white')
# data = [0,1,1,1,1,9,9,9,9,0,0]
print np.random.randn(1000)[0:10]
data = np.random.randn(1000) # store a sample from standard normal distribution of size of 1000
plt.hist(data);              # plot this smaple by histogram
```

```python
plt.hist(data, bins=30,          # number of columns         
         normed=True,            # normalize the data
         alpha=0.5,              # transparency
         histtype='stepfilled',  # generates a line plot that is by default filled
         color='steelblue',      # make the color of sttelblue  
         edgecolor='red');      # no edge color
```

```python
# take 3 samples from 3 different normal distributions, which is of the size of 1000
x1 = np.random.normal(0, 0.8, 1000)  # mean is 0, standatd deviation is 0.8
x2 = np.random.normal(-2, 1, 1000)   # mean is -2, standatd deviation is 1
x3 = np.random.normal(3, 2, 1000)    # mean is 3, standatd deviation is 2

kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40) #the same as above

plt.hist(x1, **kwargs)  # call hist function to plot the three samples
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
# combime subplots and histogram
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

```python
# using subplots to draw pie chart
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]  # store the precentage into an array
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
                          # the number shows how much the part explodes

fig1, ax1 = plt.subplots()
ax1.pie(sizes,                      # set the size
        explode=explode,            # set the explodes
        labels=labels,              # set the label
        autopct='%1.1f%%',          # set the color
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

# 6. Customizing ticks

```python
# do a subplot
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True)
```

```python
# reducing the number of ticks
# For every axis, set the x and y major locator
for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3)) # set the tick to 3
    axi.yaxis.set_major_locator(plt.MaxNLocator(3))
fig
```

```python
ax = plt.axes()
ax.plot(np.random.rand(50))

# hiding ticks or labels using NullLocator and Nullformator
ax.yaxis.set_major_locator(plt.NullLocator()) 
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')

# Set up grid, legend, and limits
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi);
```

```python
# set the fancy ticky fomat
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
fig
```

```python
# set ticks to units of pi
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```
