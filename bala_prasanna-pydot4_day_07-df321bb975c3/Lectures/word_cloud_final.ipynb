```python
from wordcloud import WordCloud

import nltk
from nltk.corpus import stopwords
s =set (stopwords.words("english"))

filename = "D:\\tmp\\nus\\bala_prasanna-pydot4_day_07-df321bb975c3\\Lectures\\input.txt"
word_dict = {}

#read the file and store the content
with open (filename) as f:
    for content in f:
        content = f.read()
        content = content.lower()

        #cleaning & sanitize
        #1. words to filter [in, ]
        
        #2. remove the punctionations
        content = content.replace(".", "")
        content = content.replace(",", "")
        content = content.replace("?", "")
        content = content.replace("!", "")

        words = content.split()

        for word in words:
            if word in word_dict:
                word_dict[word] += 1
            else:
                word_dict[word] = 1

print len(word_dict)
#print word_dict

wc = WordCloud()
wc = wc.generate_from_frequencies(frequencies=word_dict)
print (wc)

```
